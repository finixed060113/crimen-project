```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
name_year = pd.read_csv('../data/crims2010-21.csv')
```

```python
gdf_provincias = gpd.read_file('../data/geometry_prov/geom_prov.shp')
```

```python
# Crea un GeoDataFrame a partir de tus datos de infracciones
gdf_infracciones = gpd.GeoDataFrame(name_year, 
                                   geometry=gpd.points_from_xy(name_year['lng'], name_year['lat']))

# Realiza una unión espacial para encontrar a qué provincia pertenece cada punto
gdf_combined = gpd.sjoin(gdf_provincias, gdf_infracciones, how='inner', op='contains')

gdf_combined.drop(columns=['index_right'], inplace=True)
gdf_combined = gdf_combined.reset_index()
gdf_combined = gdf_combined.drop(columns=['index'])
# Ahora gdf_combined contiene la información de infracciones junto con la provincia correspondiente
```

```python
gdf_combined
```

```python
gdf_combined['name'] = gdf_combined['name'].replace('Balears (Illes), Spain', 'Islas Baleares, Spain')
gdf_combined['name'] = gdf_combined['name'].replace('Salamanca, Spain', 'Salamanca, Castilla y Leon, Spain')
gdf_combined['name'] = gdf_combined['name'].replace('Palmas (Las), Spain', 'Las Palmas de Gran Canaria, Spain')
```

```python
gdf_combined['values'] = gdf_combined['values'].astype(int)
```
